```python
import pandas as pd
import datetime 
from datetime import date, timedelta
import yfinance as yf
import plotly.graph_objects as go
import plotly.express as px
```

```python
today=date.today()

end_date=today.strftime("%Y-%m-%d")
print(end_date)
```

```python
start_date=date.today()-timedelta(days=1825)
start_date=start_date.strftime("%Y-%m-%d")
```

```python
data=yf.download("TLSA",start_date=start_date,end_date=end_date,progress=False)
data
```

```python
data["Date"]=data.index
data=data[["Date", "Open", "High", "Low", 
             "Close", "Adj Close", "Volume"]]
data.reset_index(drop=True,inplace=True)
print(data.head())
```

```python
plot=go.Figure(data=[go.Candlestick(x=data["Date"],open=data["Open"],high=data["High"],low=data["Low"],close=data["Close"])])
plot.update_layout(title = "Telsa Stock price Visualisation", xaxis_rangeslider_visible=False)
plot.show()
```

```python
plot2=px.line(data,x="Date",y="Close",title='Telsa Stock Closing price Visualisation using slider')
plot2.update_xaxes(rangeslider_visible=True)
plot2.show()
```

```python
plot3=go.Figure(data=[go.Candlestick(x=data["Date"],open=data["Open"],high=data["High"],low=data["Low"],close=data["Close"])])
plot3.update_layout(title = "Telsa Stock price Visualisation with time Period Selector", xaxis_rangeslider_visible=True)

plot3.update_xaxes(
    rangeselector=dict(
        buttons=list([
            dict(count=1,label='1m',step="month",stepmode="backward"),
            dict(count=6, label="6m", step="month", stepmode="backward"),
            dict(count=3, label="3m", step="month", stepmode="backward"),
            dict(count=1, label="1y", step="year", stepmode="backward"),
            dict(step="all")
    ])
))


plot3.show()
```
